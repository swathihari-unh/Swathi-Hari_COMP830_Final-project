```python
#importing libraries
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
from numpy import mean
from numpy import std
from numpy import dstack
from pandas import read_csv
from matplotlib import pyplot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Conv1D, LSTM, Bidirectional, GRU
from tensorflow.keras.layers import MaxPooling1D
from tensorflow.keras.utils import to_categorical
warnings.filterwarnings('ignore')
from numpy import dstack
from pandas import read_csv
import sklearn.neural_network as nn
```

```python
def load_data(): 
       
    loaded_train = ['C:/HARDataset/train/Inertial Signals/total_acc_x_train.txt', 
             'C:/HARDataset/train/Inertial Signals/total_acc_y_train.txt',
             'C:/HARDataset/train/Inertial Signals/total_acc_z_train.txt', 
          'C:/HARDataset/train/Inertial Signals/body_acc_x_train.txt',
             'C:/HARDataset/train/Inertial Signals/body_acc_y_train.txt',
             'C:/HARDataset/train/Inertial Signals/body_acc_z_train.txt',
             'C:/HARDataset/train/Inertial Signals/body_gyro_x_train.txt',
             'C:/HARDataset/train/Inertial Signals/body_gyro_y_train.txt',
            'C:/HARDataset/train/Inertial Signals/body_gyro_z_train.txt']
    
    y_train =  (read_csv('C:/HARDataset/train/y_train.txt', header=None, delim_whitespace=True)) 
    y_train = y_train.values    
    

    trainlist = list()  

    for file in loaded_train:
        data =(read_csv(file, header=None, delim_whitespace=True))
        trainlist.append(data.values) 
        train_x_data = (np.dstack(trainlist))


    
    loaded_test = ['C:/HARDataset/test/Inertial Signals/total_acc_x_test.txt', 
             'C:/HARDataset/test/Inertial Signals/total_acc_y_test.txt',
             'C:/HARDataset/test/Inertial Signals/total_acc_z_test.txt', 
          'C:/HARDataset/test/Inertial Signals/body_acc_x_test.txt',
             'C:/HARDataset/test/Inertial Signals/body_acc_y_test.txt',
             'C:/HARDataset/test/Inertial Signals/body_acc_z_test.txt',
             'C:/HARDataset/test/Inertial Signals/body_gyro_x_test.txt',
             'C:/HARDataset/test/Inertial Signals/body_gyro_y_test.txt',
            'C:/HARDataset/test/Inertial Signals/body_gyro_z_test.txt']   
  

    y_test =  (read_csv('C:/HARDataset/test/y_test.txt', header=None, delim_whitespace=True))   
    
    y_test = y_test.values     
  
    testlist = list()  
    
    for file in loaded_test:
        data =(read_csv(file, header=None, delim_whitespace=True))
        testlist.append(data.values)
        test_x_data = (np.dstack(testlist)) 
    return train_x_data, y_train, test_x_data, y_test

 
    # load the dataset, returns train and test X and y elements
    
def load_dataset():  
    #load all train
    trainX, trainy, testX, testy = load_data()
    # zero-offset class values
    trainy = trainy - 1
    testy = testy - 1
    # one hot encode y    
    trainy = to_categorical(trainy)
    testy = to_categorical(testy)
    print(trainX.shape, trainy.shape, testX.shape, testy.shape)
    print(trainX, trainy, testX, testy)
    return trainX, trainy, testX, testy
 
 
```

```python
load_dataset()
```

```python
# fit and evaluate a model
from tensorflow.keras import regularizers
def evaluate_model(trainX, trainy, testX, testy):
    verbose, epochs, batch_size = 0, 15, 32
    n_timesteps, n_features, n_outputs = trainX.shape[1], trainX.shape[2], trainy.shape[1]
    model = Sequential()
    model.add(Conv1D(filters=128, kernel_size=3, activation='relu', kernel_regularizer=regularizers.l2(0.001), input_shape=(n_timesteps,n_features)))
    model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))    
    model.add(Dropout(0.4))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dense(100, activation='relu'))
    model.add(Bidirectional(LSTM(100, input_shape=(n_timesteps,n_features), recurrent_dropout=0.4, return_sequences=True)))
    model.add(Dropout(0.4))
    model.add(Dense(100, activation='relu'))
    model.add(GRU(100, activation='tanh', return_sequences=True))
    model.add(Dropout(0.4))   
    model.add(MaxPooling1D(pool_size=4))
    model.add(GRU(100))
    model.add(Flatten())
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    # fit network
    model.fit(trainX, trainy, epochs=epochs, batch_size=batch_size, verbose=verbose)
    # evaluate model
    _, accuracy = model.evaluate(testX, testy, batch_size=batch_size, verbose=0)
    return accuracy
 
# summarize scores
def summarize_results(scores):
    print(scores)
    m, s = mean(scores), std(scores)
    print('Accuracy: %.3f%% (+/-%.3f)' % (m, s))
    
# run an experiment
def run_experiment(repeats=10):
    # load data
    trainX, trainy, testX, testy = load_dataset()
    # repeat experiment
    scores = list()
    for r in range(repeats):
        score = evaluate_model(trainX, trainy, testX, testy)
        score = score * 100.0
        print('>#%d: %.3f' % (r+1, score))
        scores.append(score)
    # summarize results
    summarize_results(scores)
    
# run the experiment
run_experiment()
```
